```python
import pandas as pd

df = pd.DataFrame({
    'windspeed': [5, 12, 8, 15],
    'temp': [20, 25, 30, 35],
    'atemp': [19, 24, 29, 34],
    'humidity': [60, 70, 65, 80]
})

df2 = pd.DataFrame({
    'windspeed': [16.6, 12.5],
    'temp': [25, 28],
    'atemp': [24, 27],
    'extra_col': ['A', 'B']
})

df
```

```python
df2
```

```python
df3 = pd.merge(df, df2, how='inner', on=['temp','atemp'])
print(df3)
```

```python
import pandas as pd

df = pd.DataFrame({
    'temp': [20, 25],
    'windspeed': [5, 12]
})

df2 = pd.DataFrame({
    'temp': [30, 35],
    'windspeed': [8, 15]
})

df3 = pd.concat([df, df2], axis=0)
print(df3)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'temp': [20, 25, 30],
    'windspeed': [5, 12, 8]
})

df2 = pd.DataFrame({
    'atemp': [19, 24, 29],
    'humidity': [60, 70, 65]
})

df3 = pd.concat([df1, df2], axis=1)
print(df3)
```
